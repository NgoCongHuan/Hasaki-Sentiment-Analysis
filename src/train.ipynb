```python
def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'\n', ' ', text)
    text = re.sub(r'\s+', ' ', text)
    text = re.sub(r'([,.!?])', r' \1 ', text)
    text = re.sub(r'\s{2,}', ' ', text)
    text = re.sub(r'[^\w\s,.!?]', '', text)
    return text.strip()

df['Nội dung đánh giá'] = df['Nội dung đánh giá'].apply(preprocess_text)
```

```python
def encode_sentiment(star_rating):
    if star_rating >= 3.0:
        return 'Tích cực'
    else:
        return 'Tiêu cực'
    
df['Cảm xúc'] = df['Số sao đánh giá'].apply(encode_sentiment)
```

```python
from pyvi import ViUtils

def add_accents(text):
    try:
        text = ViUtils.add_accents(text)
    except:
        return text
    return text

df['Nội dung đánh giá'] = df['Nội dung đánh giá'].apply(add_accents)
```
