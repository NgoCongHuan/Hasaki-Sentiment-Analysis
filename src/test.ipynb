```python
import pandas as pd
import re

# Đọc dữ liệu từ file CSV
df = pd.read_csv('../reviews.csv')

# Chuyển văn bản thành chữ thường và loại bỏ các ký tự đặc biệt
df['Nội dung đánh giá'] = df['Nội dung đánh giá'].replace('"\n"', '').replace('\r', '')
df['Nội dung đánh giá'] = df['Nội dung đánh giá'].apply(lambda x: re.sub(r'[^\w\s]', '', x.lower()))
```

```python
from vncorenlp import VnCoreNLP

def simple_usage(sentences):

    vncorenlp_file = r'../absolute/path/to/vncorenlp/VnCoreNLP-1.2.jar'

    # In this way, you have to close the server manually by calling close function
    vncorenlp = VnCoreNLP(vncorenlp_file)

    sentences_tokenizing = vncorenlp.tokenize(sentences)

    # Do not forget to close the server
    vncorenlp.close()

    return sentences_tokenizing
```

```python
# Tách từ sử dụng thư viện underthesea
df['review_tokenized'] = df['Nội dung đánh giá'].apply(simple_usage)
```

```python
def convert(nested_list):
    words = nested_list[0]
    text = " ".join(words)
    return text

df['review_tokenized'] = df['review_tokenized'].apply(convert)

df['review_tokenized']
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Tạo TF-IDF vectorizer
vectorizer = TfidfVectorizer(max_features=1000)

# Chuyển văn bản thành vector TF-IDF
X = vectorizer.fit_transform(df['review_tokenized'])
```

```python
def star(rating):
    if rating >= 3.0:
        return 1
    else:
        return -1
    
df['Số sao đánh giá'] = [1, 1, -1, 1, 1, -1, 0, 1, -1]
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

# Chia dữ liệu thành tập huấn luyện và tập kiểm tra
X_train, X_test, y_train, y_test = train_test_split(X, df['Số sao đánh giá'], test_size=0.2, random_state=42)

# Huấn luyện mô hình Logistic Regression
model = LogisticRegression()
model.fit(X_train, y_train)

# Đánh giá mô hình
accuracy = model.score(X_test, y_test)
print(f'Accuracy: {accuracy}')
```

```python
# Ví dụ: Sử dụng TF-IDF để trích xuất các từ quan trọng nhất trong các đánh giá tích cực và tiêu cực
positive_reviews = df[df['Số sao đánh giá'] == 1]['review_tokenized']
negative_reviews = df[df['Số sao đánh giá'] == -1]['review_tokenized']

positive_vectorizer = TfidfVectorizer(max_features=20)
negative_vectorizer = TfidfVectorizer(max_features=20)

positive_tfidf = positive_vectorizer.fit_transform(positive_reviews)
negative_tfidf = negative_vectorizer.fit_transform(negative_reviews)

print('Từ quan trọng trong đánh giá tích cực:', positive_vectorizer.get_feature_names_out())
print('Từ quan trọng trong đánh giá tiêu cực:', negative_vectorizer.get_feature_names_out())
```

```python
from sklearn.cluster import KMeans

# Sử dụng K-Means để phân cụm đánh giá
num_clusters = 3  # Giả định có 3 cụm
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
kmeans.fit(X)

# Gán nhãn cụm cho từng đánh giá
df['cluster'] = kmeans.labels_

# Xem các từ quan trọng trong từng cụm
for i in range(num_clusters):
    cluster_reviews = df[df['cluster'] == i]['review_tokenized']
    cluster_vectorizer = TfidfVectorizer(max_features=20)
    cluster_tfidf = cluster_vectorizer.fit_transform(cluster_reviews)
    print(f'Từ quan trọng trong cụm {i}:', cluster_vectorizer.get_feature_names_out())
```
